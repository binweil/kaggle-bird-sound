```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
## Test Sound
scale_file = "data/train_audio/asbfly/XC134896.ogg"
ipd.Audio(scale_file)
```

```python
# load audio files with librosa
scale, sample_rate = librosa.load(scale_file)
```

```python
# n_fft: frame rate, sr: sampleing rate, n_mels: number of mel bands
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
filter_banks.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(filter_banks, sr=sample_rate, x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mel_spectrogram = librosa.feature.melspectrogram(y=scale, n_fft=2048, sr=22050, n_mels=10, hop_length=512)
mel_spectrogram.shape
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
log_mel_spectrogram.shape
```

```python
plt.figure(figsize=(25,10))
librosa.display.specshow(log_mel_spectrogram, x_axis="time", y_axis="mel", sr=sample_rate)
plt.colorbar(format="%+2.f")
plt.show()
```
